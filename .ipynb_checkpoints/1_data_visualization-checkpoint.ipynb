# Data Visualization


This notebook will be intended to visualize the features present within our dataset. Each graph visualization will be appropriately labeled and will have an attached description to clarify the intent of the graph as well as a short description that details what information was gained.

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
df_fundamentals = pd.read_csv('./data/fundamentals.csv')
df_prices = pd.read_csv('./data/prices.csv')
df_prices_adj = pd.read_csv('./data/prices-split-adjusted.csv')
df_securities = pd.read_csv('./data/securities.csv')
```

Below are line-graph visualizations of 5 stocks' prices against time - from 2010 to 2016. We start with Willis Towers' graph.

```python
# df_prices.groupby('symbol').plot(x='date', y='high')
```

```python
df = df_prices.loc[df_prices['symbol'] == 'WLTW']
# df_prices.loc[df_prices['symbol'] == 'WLTW'].plot(x='date', y='low')
df
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(10,8))
ax.plot(df['date'], df['high'],label='DAILY HIGH')
ax.plot(df['date'], df['low'], c='r', label='DAILY LOW')
ax.set_title('Willis Towers Watson PLC')
ax.set_ylabel('Stock Price')
ax.set_xlabel('Time (2010-2016)')
ax.tick_params(axis='x',
               which='both',
               bottom=False,
               top=False,
               labelbottom=False)
ax.legend()
# ax.show()
```

Similarly for the other four stocks:

```python
stocks ={'AAPL': 'Apple Inc', 'XOM': 'ExonMobil Corporation','AAL': 'American Airlines Group Inc','CVS': 'CVS Health Group'}
for ticker, name in stocks.items():
    df = df_prices.loc[df_prices['symbol'] == ticker]
    # df_prices.loc[df_prices['symbol'] == 'WLTW'].plot(x='date', y='low')
    df
    fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(10,8))
    ax.plot(df['date'], df['high'],label='DAILY HIGH')
    ax.plot(df['date'], df['low'], c='r', label='DAILY LOW')
    ax.set_title(name)
    ax.set_ylabel('Stock Price')
    ax.set_xlabel('Time (2010-2016)')
    ax.tick_params(axis='x',
                   which='both',
                   bottom=False,
                   top=False,
                   labelbottom=False)
    ax.legend()
```

Note that the dramatic apparent decrease in Apple's stock price was caused by a 7-1 stock split in 2014.
