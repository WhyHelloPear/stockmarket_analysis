# Data Visualizations
Below are line-graph visualizations of 5 stocks' prices against time - from 2010 to 2016. We start with Willis Towers' graph.

```python
import numpy as np
import pandas as pd
import math
import sklearn
import sklearn.preprocessing
import datetime
import os
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df_fundamentals = pd.read_csv('./data/fundamentals.csv')
df_prices = pd.read_csv('./data/prices.csv')
df_securities = pd.read_csv('./data/securities.csv')
```

```python
df = df_prices.loc[df_prices['symbol'] == 'WLTW']

fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(20,10))
ax.plot(df['date'], df['high'],label='DAILY HIGH')
ax.plot(df['date'], df['low'], c='r', label='DAILY LOW')
ax.set_title('Willis Towers Watson PLC Price History')
ax.set_ylabel('Stock Price')
ax.set_xlabel('Time (2010-2016)')
ax.tick_params(axis='x',
               which='both',
               bottom=False,
               top=False,
               labelbottom=False)
ax.legend()
```

```python
df = df_prices.loc[df_prices['symbol'] == 'WLTW']

fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(20,10))
ax.plot(df['date'], df['volume'], c='darkorange')
ax.set_title('Willis Towers Watson PLC Volume History')
ax.set_ylabel('Stock Volucme')
ax.set_xlabel('Time (2010-2016)')
ax.tick_params(axis='x',
               which='both',
               bottom=False,
               top=False,
               labelbottom=False)
```

Similarly for the other four stocks:

Note that the dramatic apparent decrease in Apple's stock price was caused by a 7-1 stock split in 2014.

```python
tickers = ['AAPL','XOM','AAL','CVS','WLTW']
names = ['Apple Inc','ExonMobil Corporation','American Airlines Group Inc','CVS Health Group','Willis Towers Watson PLC']

fig, ax = plt.subplots(5,2, figsize = (20,30))

i = 0
j = 0
for k in range(len(tickers)*2):
    if j > 1:
        j = 0
        i += 1
    if j == 0:
        df = df_prices.loc[df_prices['symbol'] == tickers[int(k/2)]]
        ax[i][j].plot(df['date'], df['high'],label='DAILY HIGH')
        ax[i][j].plot(df['date'], df['low'], c='r', label='DAILY LOW')
        ax[i][j].set_title(names[int(k/2)]+" Price History")
        ax[i][j].set_ylabel('Stock Price')
        ax[i][j].set_xlabel('Time (2010-2016)')
        ax[i][j].tick_params(axis='x',
                       which='both',
                       bottom=False,
                       top=False,
                       labelbottom=False)
        ax[i][j].legend()
        
    else:
        df = df_prices.loc[df_prices['symbol'] == tickers[int(k/2)]]
        ax[i][j].plot(df['date'], df['volume'], c='darkorange')
        ax[i][j].set_title(names[int(k/2)]+" Volume History")
        ax[i][j].set_ylabel('Stock Volume')
        ax[i][j].set_xlabel('Time (2010-2016)')
        ax[i][j].tick_params(axis='x',
                       which='both',
                       bottom=False,
                       top=False,
                       labelbottom=False)

    j += 1
```
